```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

# 2차원 배열로 만들기
fish_data = [[l,w] for l, w in zip(fish_length, fish_weight)]
# 결과데이터 준비
fish_target = [1]*35+[0]*14

print(fish_data)
print(fish_target)
```

```python
# 랜덤으로 배열 섞어주고 그래프로 출력
import numpy as np

input_arr = np.array(fish_data) # 입력 49개
target_arr = np.array(fish_target)  # 타겟 49개

print(input_arr.shape)
print(target_arr.shape)
```

```python
# 랜덤을 섞기 전 그래프 출력(기준 : 35)
import matplotlib.pyplot as plt

plt.scatter(input_arr[:35, 0], input_arr[:35, 1])
plt.scatter(input_arr[35:, 0], input_arr[35:, 1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# ramdom으로 배열 섞기
index = np.arange(49)   #0~48
np.random.shuffle(index)

print(index)
# [ 3 40 46 23 41 19  7 29 48 43 35 31 36 28 44  6 26  5 15 20  0 45 30 10 13 42 39  4 34 18 24 22  8 38 37 14 32 17 12 33  1 16 11 25 47  9 21 27 2]
# [35  6 46  1  4 32 43 45 33 21 37 26 20 27 19 13  2  0 29 12 15 42 34  7 16 36 23 24 44 17 10 28  8 31  9 40 38 14 18 30 11 48  3 39  5 41 25 47 22]
```

```python
# random으로 섞인 배열로 훈련데이터 생성
train_input = input_arr[index[:35]] # 35개 데이터(입력)
train_target = target_arr[index[:35]]   # 35개 데이터(타겟)
test_input = input_arr[index[35:]]  # 14개 데이터(평가 - 입력)
test_target = target_arr[index[35:]]    # 14개 데이터(평가 - 타겟)
```

```python
# 랜덤을 섞은 후 그래프 출력(기준 : 35)
plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 훈련시작
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
kn.predict(test_input)
test_target
```
